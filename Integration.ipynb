```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt

```

```python
#Define Functions

def f1(x):
    return np.sin(x)

def f2(x):
    return np.cos(x)

def f3(x):
    return np.exp(-x)
```

```python

x_vals = np.linspace(0, np.pi, 100)
x_vals_exp = np.linspace(0, 2, 100)

# Visualise
plt.figure(figsize=(12, 6))

# Plot sin(x)
plt.subplot(1, 3, 1)
plt.plot(x_vals, f1(x_vals), label='sin(x)', color='b')
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.fill_between(x_vals, 0, f1(x_vals), color='b', alpha=0.1)

# Plot cos(x)
plt.subplot(1, 3, 2)
plt.plot(x_vals, f2(x_vals), label='cos(x)', color='g')
plt.title('cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.fill_between(x_vals, 0, f2(x_vals), color='g', alpha=0.1)

# Plot exp(-x)
plt.subplot(1, 3, 3)
plt.plot(x_vals_exp, f3(x_vals_exp), label='exp(-x)', color='r')
plt.title('exp(-x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.fill_between(x_vals_exp, 0, f3(x_vals_exp), color='r', alpha=0.1)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Compute integrals
result_f1, error_f1 = integrate.quad(f1, 0, np.pi)
result_f2, error_f2 = integrate.quad(f2, 0, np.pi)
result_f3, error_f3 = integrate.quad(f3, 0, 2)
```

```python
# Display integral results
print(f"Integral of sin(x) from 0 to pi: {result_f1}, Error: {error_f1}")
print(f"Integral of cos(x) from 0 to pi: {result_f2}, Error: {error_f2}")
print(f"Integral of exp(-x) from 0 to 2: {result_f3}, Error: {error_f3}")
```
